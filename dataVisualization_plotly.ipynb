```python
import pandas as pd
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("MARKET_Car_Prices_clean.csv")
```

```python
df
```

```python
fig = px.scatter(df, x="price", y="city_mpg", hover_name="make", size="engine_size", color="make")
```

```python
fig
```

```python
fig = px.scatter(df, x="engine_size", y="horsepower", color="fuel_type", marginal_y="violin",
           marginal_x="box", trendline="ols", template="simple_white")
fig.show()
```

```python
labels = df['make'].unique()
values = [df["make"].value_counts()[x] for x in labels]
```

```python
data = go.Pie(labels=labels, values=values, hole=0.6)

layout = {"title": "Percentage of car makes"}

go.Figure(data=data, layout=layout)
```

```python
go.Figure(go.Bar(y=values, x=labels,))
```

```python
mean_city_mpg = [df.loc[df['make'] == x, 'city_mpg'].mean() for x in labels]
mean_highway_mpg = [df.loc[df['make'] == x, 'highway_mpg'].mean() for x in labels]
```

```python
trace_1 = go.Bar(x=labels, y=mean_city_mpg, name="Mean city mpg")
trace_2 = go.Bar(x=labels, y=mean_highway_mpg, name="Mean highway mpg")

data = [trace_1,trace_2]

fig = go.Figure(data)
fig.show()
```

```python
data = go.Histogram(x=df['price'])

layout = dict(title="Histogram of price")

fig = go.Figure(data, layout)
fig.show()
```
