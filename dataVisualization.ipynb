Corelation between num_of_cylinders and highway_mpg

```python
df['num_of_cylinders'].unique()
```

```python
# replace string values to int for num_of_cylinders
df['num_of_cylinders'] =df['num_of_cylinders'].map({"two":2, "three":3, "four":4,"five":5,"six":6, "eight":8, "twelve":12})
```

```python
df['num_of_cylinders'].unique()
```

```python
df['num_of_cylinders'].corr(df['highway_mpg'])
```

```python
plt.scatter(df['num_of_cylinders'], df['highway_mpg'],marker="<", color="b", label="highway")
plt.scatter(df['num_of_cylinders'], df['city_mpg'],marker=">", color="r", label="city")
plt.xlabel("num_of_cylinders")
plt.ylabel("miles_per_galon")
plt.legend()
plt.show()
```
